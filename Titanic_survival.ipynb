```python
# import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/archive.zip")
df
```

```python
# basic info
df.info()
```

```python
# missing values check
df.isnull().sum()
```

```python
# clean missing values
df["Age"]=df["Age"].fillna(df["Age"].median())
df["Embarked"]=df["Embarked"].fillna(df["Embarked"].mode()[0])

```

```python
# EDA Visualizations
# Survival count
sns.countplot(x='Survived', data=df)
plt.title("Survival Count")
plt.show()
```

More people did not survive compared to those who survived.
Survival rate is low

```python
# Gender Distribution
sns.countplot(x='Sex', data=df)
plt.title("Gender Count")
plt.show()
```

More male passengers than females on the ship

```python
#Survival by Gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title("Survival by Gender")
plt.show()
```



• Females survived more than males.



---


```python
#Age Distribution
df['Age'].plot(kind='hist', bins=30, figsize=(8,4))
plt.title("Age Distribution")
plt.xlabel("Age")
plt.show()
# Majority travelled in 3rd Class (cheapest). • Few people in 1st Class (rich passengers).
```

 Most passengers were between 20–40 years. • Very few older passengers (above 60).




```python

# Passenger Class Count
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
#Survival by Passenger Class
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title("Survival Rate by Class")
plt.show()
```

• 1st class passengers survived more than 2nd and 3rd class. • 3rd class survival is very low.

```python
#Fare distribution
df['Fare'].plot(kind='hist', bins=30)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.show()
```




• Most fares are low (cheaper tickets).
• Few very expensive tickets → mostly 1st class.



```python
#Correlation Heatmap
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(8,6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['Sex'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Gender Percentage")
plt.ylabel("")
plt.show()
```
